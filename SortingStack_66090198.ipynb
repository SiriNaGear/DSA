```python
class Node:
  '''Class for storing a singly linked node.'''
  def __init__(self, element, next):
    self._element = element
    self._next = next

  def __str__(self):
    return str(self._element)

class Linked_Queue:
    '''
    FIFO queue implementation using a singly linked list for storage.
    '''
    def __init__(self):
        '''Create an empty queue.'''
        self._header = None
        self._trailer = None
        self._size = 0

    def __len__(self):
        '''Return the number of elements in the queue.'''
        return self._size

    def __str__(self):
        data = []
        point = self._header
        while point:
            data.append(str(point))
            point = point._next
        return ', '.join(data)

    def is_empty(self):
        '''Return True if the queue is empty'''
        return self._size == 0

    def first(self):
        '''Return the element at the front of the queue.'''
        if self.is_empty():
            raise Exception('Queue is empty')
        return self._header._element

    def dequeue(self):
        '''
        Remove and return the first element of the queue.
        Raise Exception if the queue is empty
        '''
        if self.is_empty():
            raise Exception('Queue is empty')
        data = self._header._element
        self._header = self._header._next
        self._size -= 1
        if self.is_empty():
            self._trailer = None
        return data

    def enqueue(self, x):
        '''Add an element to the back of queue.'''
        newest = Node(x, None)
        if self.is_empty():
            self._header = newest
        else:
            self._trailer._next = newest
        self._trailer = newest
        self._size += 1

    def push(self, x):
        self._data.append(x)

    def pop(self):
        if self.is_empty():
            raise Exception('Stack is empty')
        return self._data.pop()

    def top(self):
        if self.is_empty():
            raise Exception('Stack is empty')
        return self._data[-1]
```
